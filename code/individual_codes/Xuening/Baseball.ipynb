```python
print(Team19.head())
select_data = Team19[['Rk', 'Tm']]
select_data.dropna(inplace=True)
select_data
```

```python
# Assign the Tiers
# Assign tiers based on the 'Rk' (rank)
# For example, 1-5 is Tier 1, 6-10 is Tier 2, etc.
def assign_tier(rank):
    if 1 <= rank <= 7:
        return 1
    elif 8 <= rank <= 14:
        return 2
    elif 15 <= rank <= 21:
        return 3
    elif 22 <= rank <= 30:
        return 4

# Apply the function to the 'Rk' column to create a new 'Tier' column
select_data['Tier'] = select_data['Rk'].apply(assign_tier)

Tierrank = select_data[['Rk', 'Tm', 'Tier']]  # Display the DataFrame with the new 'Tier' column
Tierrank
```

```python
import pandas as pd
import plotly.express as px
import plotly.graph_objs as go

# Handling NaN values by replacing them with the mean of the WHIP column
final_combined['WHIP'].fillna(final_combined['WHIP'].mean(), inplace=True)

# Filter the DataFrame to include only Washington Nationals' data for emphasis.
nationals_df = final_combined[final_combined['Abbreviation'] == 'WNS']

# Create an interactive plot for all teams.
fig = px.scatter(final_combined, x='HR', y='SO9', size='WHIP', color='Abbreviation',
                 hover_name='Name',  # Use 'Name' for more detailed hover information.
                 title='Pitching Analysis for All Teams with Emphasis on Washington Nationals')

# Add Washington Nationals' data as a separate trace to emphasize it.
nationals_trace = go.Scatter(x=nationals_df['ERA'], y=nationals_df['SO9'],
                             mode='markers+text', text='WNS', textposition='top center',
                             marker=dict(color='red', size=nationals_df['WHIP']*20,
                                         line=dict(width=2)),
                             name='Washington Nationals')

fig.add_trace(nationals_trace)

# Show the plot.
fig.show()

```

```python
final_combined_cleaned = final_combined.dropna()
final_combined_cleaned = final_combined_cleaned[final_combined_cleaned['ERA']<=20]

# Group by team and select the top 3 players by games played
top3_by_team = final_combined_cleaned.groupby('Tm_x').apply(lambda x: x.nlargest(3, 'G')).reset_index(drop=True)
```

```python


# Create an interactive plot using logos as hover images (hovertext)
fig = px.scatter(
    top3_by_team,
    x='ERA', 
    y='SO9', 
    size='HR',  # Make sure this column has no NaN values
    color='Tier',  # Coloring by 'Tier' to differentiate groups
    hover_name='Full_Name',  # Shows the full team name on hover
    hover_data='Logo',  # Show logo URLs on hover
    title='Pitching Analysis for All Teams'
)

# Display the plot
fig.show()
```
